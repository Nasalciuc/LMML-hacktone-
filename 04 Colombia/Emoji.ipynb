```python
import os
import pandas as pd

possible_paths = [
    "participant_input/input.txt",
    "../participant_input/input.txt",
    "./input.txt"
]

# Find the first existing path
for path in possible_paths:
    if os.path.exists(path):
        input_path = path
        break
else:
    raise FileNotFoundError("❌ Could not find input.txt in participant_input or current folder.")

# Read and process
with open(input_path, "r", encoding="utf-8") as f:
    lines = f.read().splitlines()

df = pd.DataFrame({"line": lines})
df["cleaned"] = df["line"].str.replace(".", "", regex=False)

# Save output
with open("output.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(df["cleaned"]))

print("✅ output.txt created successfully!")
```
